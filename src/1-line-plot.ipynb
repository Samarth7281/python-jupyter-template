### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x and y values
x = np.arange(0, 11)       # x from 0 to 10
y = x ** 2                 # y = x squared

# Create the plot
plt.plot(x, y, marker='o', color='blue', linestyle='-')

# Customize the plot
plt.title("Line Plot of y = x²")
plt.xlabel("x values")
plt.ylabel("y values (x²)")
plt.grid(True)

# Show the plot
plt.show()
```
